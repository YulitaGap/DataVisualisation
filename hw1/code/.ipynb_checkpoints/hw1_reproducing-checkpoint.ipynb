# HW 1 Data Visulisation
## Reproducing NYT chart

```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("../data/data.csv",skiprows=1)
df.head()
```

### Data preparation
To reproduce a NYT chart we need to use a subset of the monthly means data. 
The input file also includes the columns J-D, D-N, DJF, MAM, JJA and SON, which from my research stand for different seasons annual means, but I decided to drop those columns and make calculations using raw monthly data.

```python
# Deleting extra columns

df.drop(columns=df.columns[-6:], 
        axis=1, 
        inplace=True)
df.head()
```

Now we need to calculate the average annual value based on monthly values.

```python
#Adding avg column

df = df.assign(avg=df.iloc[:,1:13].mean(axis=1))
df
```

To draw a line which measures hotter/colder then 1880-1899 years let's calculate that value.

```python
#Creating a horizontal line points to visualize 1880-1899 avg temp value

line_avg = df.loc[df['Year'] <= 1890]["avg"].mean()

line_df = pd.DataFrame({
    'Year': [1880,2000,2020],
    'avg': [line_avg, line_avg,line_avg],
    'title' : ["", "HOTTER THEN 1880-1899 AVERAGE", "COLDER"]
})
```

Now starting to work on chart

```python
chart = alt.Chart(df).properties( width = 950, height = 600,
             title= { "text": ["Annual Global Surface Temperature, ", "Relative to Late 19th Century Average"]})


mark = chart.mark_point(filled = True, size = 70,color = 'orangered', strokeWidth = 1, stroke='black').encode(
    x = alt.X("Year:Q", title=None, axis=alt.Axis(tickCount=10, format="d")), 
    y = alt.Y("avg:Q", title=None, axis=alt.Axis(tickCount=10, format="+20"), scale=alt.Scale(domain=[-0.6, 1.2]))
)


line = alt.Chart(line_df).mark_line(color= 'grey', strokeWidth=1).encode(
    x= 'Year',
    y= 'avg',
)

line_text_hotter = line.mark_text(size = 15, color = 'black',dy=-20).encode(
    text = 'title'
).transform_filter(
    alt.FieldEqualPredicate(field='Year', equal=2000))

line_text_colder = line.mark_text(size = 15, color = 'black',dy=20, dx=-50).encode(
    text = 'title'
).transform_filter(
    alt.FieldEqualPredicate(field='Year', equal=2020))

res = (line+line_text_hotter+line_text_colder+mark).configure_axisLeft(labelColor='grey').configure_axisBottom(labelColor='grey').configure_title(
    fontSize=20,
    font='Arial',
    anchor='start',
    dx = 50,
    dy = 70,
    color='Black'
)
res
```

We also need to highlight some years values on graph. (I didn't find out if they are selected for some property, so just a dump check.)

```python
#Creating additional column to mark years which need to be hightlighted on chart

df['highlight'] = (df['Year'] == 1904) | (df['Year'] == 1944) | (df['Year'] == 1998) | (df['Year'] == 2014) | (df['Year'] == 2015) | (df['Year'] == 2016) | (df['Year'] == 2017)
df
```

```python
#Adding a labels for highlighted years.

text = (alt.Chart(df).mark_text(size = 15, dx=-25, color="black")
    .encode(x=alt.X("Year"), 
            y=alt.Y("avg"),
            text = alt.Text('Year')
).transform_filter(
    alt.FieldEqualPredicate(field='highlight', equal=True)))

(res+text).configure_axis(
    labelFontSize=14,
    labelFont = "Arial"
)
```
