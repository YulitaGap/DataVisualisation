```python
import pandas as pd
import altair as alt
import geopandas as gpd
```

```python
#Reading data
df = pd.read_excel("../data/data.xlsx")
df.head()
```

```python
df = pd.melt(df, id_vars=['Регіон, рік'], 
                  var_name="field", value_name="percent")

df.rename(columns={'Регіон, рік':'region_year'}, inplace=True)
df[['region','year']] = df.region_year.str.split(",",expand=True,)
df.head()
```

```python
from IPython.display import display, HTML

display(HTML("""
<style>
form.vega-bindings {
  position: absolute;
  left: 0px;
  top: 0px;
}
</style>
"""))
```

```python
input_dropdown = alt.binding_select(options = df.region.unique(), name='Регіон ')
selection = alt.selection_single(fields = ['region'], bind = input_dropdown)


alt.Chart(df).mark_bar().add_selection(
    selection
).encode(
    x=alt.X('sum(percent):Q', stack="normalize", axis=alt.Axis(format='%'), title='Відстоткова частка'),
    y=alt.Y('region_year', title=''),
    color= alt.Color('field', title="Напрями"),
    tooltip=[
        alt.Tooltip('year', title='Рік'), 
        alt.Tooltip('field', title='Напрям'), 
        alt.Tooltip('percent', title='%')]
).transform_filter(
    selection
).properties( width = 700,
             title= { "text": ["Зміна структури ВДВ у розрізі регіонів ", "за 2017 порівняно з 2010"]})
```
